```python
#900都市のマッチング描画

import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance_matrix
from scipy.optimize import linear_sum_assignment

def read_tsp(file_path):
    """ TSP LIB 形式のファイルから頂点情報を読み取る """
    nodes = {}
    with open(file_path, 'r') as f:
        lines = f.readlines()
    
    is_coord_section = False
    for line in lines:
        line = line.strip()
        if line.startswith("NODE_COORD_SECTION"):
            is_coord_section = True
            continue
        if line.startswith("EOF"):
            break
        if is_coord_section:
            parts = line.split()
            if len(parts) >= 3:
                node_id = int(parts[0])
                x, y = float(parts[1]), float(parts[2])
                nodes[node_id] = (x, y)
    return nodes

def extract_unique_nodes(nodes1, nodes2):
    """ 2つのノードセットから片方にしか存在しないノードを抽出 """
    set1 = set(nodes1.keys())
    set2 = set(nodes2.keys())
    
    unique1 = {k: nodes1[k] for k in set1 - set2}
    unique2 = {k: nodes2[k] for k in set2 - set1}
    
    return unique1, unique2

def solve_assignment(unique1, unique2):
    """ 最小重み完全マッチングをハンガリー法で解く """
    ids1, coords1 = zip(*unique1.items()) if unique1 else ([], [])
    ids2, coords2 = zip(*unique2.items()) if unique2 else ([], [])
    
    if not coords1 or not coords2:
        return {}  # 一方が空ならマッチング不可
    
    cost_matrix = distance_matrix(coords1, coords2)
    row_ind, col_ind = linear_sum_assignment(cost_matrix)
    
    matches = {ids1[i]: ids2[j] for i, j in zip(row_ind, col_ind)}
    return matches

def plot_matching(unique1, unique2, matches, output_file=None):
    """ 一致しないノードとそのマッチングを可視化する（座標軸なし） """
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_facecolor('white')
    
    # ノードをプロット
    x1, y1 = zip(*unique1.values()) if unique1 else ([], [])
    x2, y2 = zip(*unique2.values()) if unique2 else ([], [])

    ax.scatter(x1, y1, color='black',  alpha=0.7, s=1)
    ax.scatter(x2, y2, color='red',  alpha=0.7, s=1)

    # マッチングを線で結ぶ
    for node1, node2 in matches.items():
        x_start, y_start = unique1[node1]
        x_end, y_end = unique2[node2]
        ax.plot([x_start, x_end], [y_start, y_end], color='black', linestyle='--', linewidth=1, alpha=0.7)

    # ラベルと凡例（軸は非表示）
    ax.legend(frameon=False, loc='upper right', fontsize=10)
    ax.axis('off')  # 座標軸を削除
    
    # 画像として保存する場合
    if output_file:
        plt.savefig(output_file, dpi=300, bbox_inches='tight', pad_inches=0.1)
    
    plt.show()

def main(tsp_file1, tsp_file2, output_image=None):
    """ 2つの TSP ファイルを受け取り、マッチング結果を可視化 """
    nodes1 = read_tsp(tsp_file1)
    nodes2 = read_tsp(tsp_file2)
    
    unique1, unique2 = extract_unique_nodes(nodes1, nodes2)
    matches = solve_assignment(unique1, unique2)
    
    plot_matching(unique1, unique2, matches, output_image)

# 使用例
tsp_file1 = "original_90k_n1.tsp"
tsp_file2 = "similar_90K_n1.tsp"
output_image = "matching_visualization.png"  # 画像として保存（不要なら None）

main(tsp_file1, tsp_file2, output_image)

```
